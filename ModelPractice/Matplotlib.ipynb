## **Import**


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets
#加上該行就不用每次都要show()
%matplotlib inline
```

## **Matplotlib**


```python
# x預設0,1,2,3
plt.plot([1, 2, 3, 4])
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 17], "ro")
plt.xlabel("x")
plt.ylabel("y")
```

```python
t = np.arange(0., 5., 0.2)

# y = x, red dashes
# y = x^2, blue squares
# y = x^3, green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
'''
第一對'a'是從0到49的整數陣列
第二對'c'是長度為50且範圍在0到49之間（不包括50）的隨機整數陣列
第三對'd'是長度為50並符合標準正態分佈（平均值為0，方差為1）的隨機浮點數陣列
'''
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a']+10*np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

# x index, y index, color, size
plt.scatter('a', 'b', c='c', s='d', data=data)
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 9))

plt.subplot(331)  # (3行3列第一個)
plt.bar(names, values)
plt.subplot(332)  # (3行3列第二個)
plt.scatter(names, values)
plt.subplot(339)  # (3行3列第九個)
plt.plot(names, values, linewidth='7')
plt.suptitle('Categorical Plotting')
```

## **Pandas+Matplotlib**


```python
iris = datasets.load_iris()
x = pd.DataFrame(iris['data'], columns=iris['feature_names'])
print("target_names: "+str(iris['target_names']))
y = pd.DataFrame(iris['target'], columns=['target'])
iris_data = pd.concat([x, y], axis=1)
iris_data.head()
```

```python
iris_data.plot.scatter(x='sepal length (cm)', y='sepal width (cm)', c='target')
```

```python
color = {
    0: 'r',
    1: 'g',
    2: 'b'
}
iris_data['color'] = iris_data['target'].map(color)
iris_data.head(5)
iris_data.plot.scatter(x='sepal length (cm)',
                       y='sepal width (cm)', c=iris_data['color'])
```

```python
ax = iris_data[iris_data['target'] == 0].plot.scatter(
    x='sepal length (cm)', y='petal width (cm)', color='r', label='setosa')
ax = iris_data[iris_data['target'] == 1].plot.scatter(
    x='sepal length (cm)', y='petal width (cm)', color='g', label='versicolor', ax=ax)
iris_data[iris_data['target'] == 2].plot.scatter(
    x='sepal length (cm)', y='petal width (cm)', color='b', label='virginica', ax=ax)
```
