## **Import Package**


```python
import pandas as pd
from sklearn import datasets
import seaborn as sns
```

## **Load Data**


```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
planets = sns.load_dataset("planets")
```

## **Tips**


```python
tips
```

```python
sns.scatterplot(x="tip", y="total_bill", data=tips,
                hue="day", size="size", palette="YlGnBu")
```

```python
# kde:曲線 bins:條柱數
sns.histplot(data=tips, x="tip", kde=True, bins=20)
```

```python

sns.displot(data=tips, x="tip", kde=True, bins=20)
```

```python
sns.barplot(x="sex", y="tip", data=tips, palette="YlGnBu")
```

```python
sns.boxplot(x="day", y="total_bill", data=tips, hue="sex", palette="YlGnBu")
```

```python
sns.stripplot(x="day", y="tip", data=tips, hue="sex",
              palette="YlGnBu", dodge=True)
```

```python
sns.jointplot(x="tip", y="total_bill", data=tips, kind="reg")
```

```python
sns.jointplot(x="tip", y="total_bill", data=tips, kind="kde", shade=True)
```

```python
sns.jointplot(x="tip", y="total_bill", data=tips, kind="hex")
```
